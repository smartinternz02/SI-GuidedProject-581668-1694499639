## Task 2.2 -> Load the dataset.

```python
# Importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Loading the dataset
df = pd.read_csv('/content/House Price India.csv')
df.head()
```

## Task 2.3 ->  Perform the Below Visualizations.
* Uni - variate Analysis
* Bi - variate Analysis
* Multi - variate Analysis

### Uni-variate Analysis

```python
# Scatter plot for the 'Lattitude' column
plt.figure(figsize=[15,15])

sns.scatterplot(df['Lattitude'])

plt.title('Lattitude of each houses')
```

```python
# Line plot for the 'Price' column
plt.figure(figsize=[15,15])

sns.lineplot(df['Price'])

plt.title('Price of each house')
```

```python
# Pie chart for the 'Number of schools nearby' column
plt.pie(
    df.head(10)['Number of schools nearby'],
    labels = df.head(10)['id'],
    autopct = '%.1f%%'
)

plt.title('Number of Schools nearby - for the first 10 houses')
```

### Bi-variate Analysis

```python
# Segmented Bar graph of 'Longitude' and 'Lattitude' w.r.t the 'id'
plt.figure(figsize = [15,15])

plt.bar(df['id'], df['Lattitude'], label="Lattitude", width=0.5, color='y')
plt.bar(df['id'], df['Longitude'], label="Longitude", width=0.5, color='g')

plt.xlabel("ID of the house")
plt.ylabel("Longitude & Lattitude")
plt.legend()
plt.title('Lattitude & Longitude for each house')
```

```python
# Lineplot of 'living_area_renov' of each house
plt.figure(figsize = [15,15])

sns.lineplot(
    x=df['id'],
    y=df['living_area_renov']
)

plt.xlabel("ID of the house")
plt.ylabel("Living Area Renov")
plt.title('Living Area')
```

### Multi-variate Analysis

```python
# Heatmap
plt.figure(figsize = [20,20])

sns.heatmap(df.corr(), annot=True)
```
